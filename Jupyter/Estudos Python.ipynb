```python
def save(nome, valor):
    num = range(1000)
    json = [{
        "id": num,
        "nome": nome,
        "valor": valor,
    }]
    with open("text.txt", "w") as f:
        f.write(f"{json} \n")
        print("Arquivo criado com sucesso!")
```

```python
def read():
    if os.path.exists("text.txt"):
        with open("text.txt", "r") as f:
            return f.read()

print(read())
```

```python
class Product:
    produtos = []
    
    def __init__(self, nome, preco):
        self._nome = nome
        self._price = preco
                
        self.carregar_dados()

    # Get Nome
    def getNome(self):
        return self._nome

    # Get Price
    def getPrice(self):
        return self._price

    def getProdutos(self):
        for produto in self.produtos:
            return produto

    def setProdutos(self):
        self.produtos.append(f"{self.getNome()} - {self.getPrice()}") 
        return (f"Produto {self.getNome()} cadastrado com sucesso.")
        
    def resumir(self):
        return f"{self.getNome()} custa R$ {self.getPrice()}"

    def carregar_dados(self):
        if os.path.exists("text.txt"):    
            if len(self.produtos) == 0:
                with open("text.txt", "r") as f:
                    self.produtos.append(f.read())
            return self.produtos
        else:
            with open("text.txt", "w") as f:
                f.close()
        
```

```python
#instacia do objeto
def product(name, price):
    p1 = Product(name, price)
    p1.nome = name
    p1.price = price
    p1.setProdutos()
    return p1
```
