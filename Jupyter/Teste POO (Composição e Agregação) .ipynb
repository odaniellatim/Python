```python
class Clientes:
    def __init__(self,id_cliente, nome_cliente, email_cliente, telefone_cliente, interesses_cliente):
        self.id_cliente = id_cliente
        self.nome_cliente = nome_cliente
        self.email_cliente = email_cliente
        self.telefone_cliente = telefone_cliente
        self.interesses_cliente = interesses_cliente

    def get_nome(self):
        return self.nome_cliente

    def get_email(self):
        return self.email_cliente

    def get_interesses(self, nome_cliente):
        interesses_cliente = ", ".join(self.interesses_cliente)
        print(f"Cliente: {self.nome_cliente.ljust(50,'.')} Interesses: {interesses_cliente}")

    def get_info_clientes(self):        
        interesses_cliente = ", ".join(self.interesses_cliente)
        print(f"")
        print(f"Nome Cliente: {self.nome_cliente.ljust(50,'.')} ID: {self.id_cliente}")
        print(f"E-mail Cliente: {self.email_cliente.rjust(50,'.')}")
        print(f"Telefone Cliente: {self.telefone_cliente.rjust(50,'.')}")
        print(f"Interesses do Cliente: {interesses_cliente.rjust(50,'.')}")
        print(f"")

    def cliente_atualizar_info(self, id_cliente, nome_cliente, email_cliente, telefone_cliente, interesses_cliente):
        self.nome_cliente = nome_cliente
        self.email_cliente = email_cliente
        self.telefone_cliente = telefone_cliente
        self.interesses_cliente = interesses_cliente
        print(f"Cliente {nome_cliente} atualizado com sucesso!")
    
        
```

```python
import os
import json

class Locadora:
    def __init__(self):
        self.clientes =  []

# Carrega a base de dados que está sendo usada como um dicionario.
    # def bd(self, base_de_dados=None):
    #     if base_de_dados:
    #         for id, db in base_de_dados.items():
    #             for cliente in db:
    #                 id_cliente = cliente['id']
    #                 nome_cliente = cliente['nome']
    #                 email_cliente = cliente['email']
    #                 telefone_cliente = cliente['telefone']
    #                 interesses_cliente = cliente['interesses']

    #                 instacia_clientes = Clientes(id_cliente, nome_cliente, email_cliente, telefone_cliente, interesses_cliente)
    #                 self.clientes.append(instacia_clientes)
    #             print("Base de dados carregado com sucesso!")
    #     else:
    #         print("Base de dados não informada.")

    def carregar_dados_do_json(self, nome_arquivo):
        if os.path.exists(nome_arquivo):

            with open(nome_arquivo, 'r', encoding='utf-8') as arquivo_json:
                base_de_dados = json.load(arquivo_json)
                for cliente in base_de_dados['clientes']:
                    id_cliente = cliente['id']
                    nome_cliente = cliente['nome']
                    email_cliente = cliente['email']
                    telefone_cliente = cliente['telefone']
                    interesses_cliente = cliente['interesses']

                    instancia_clientes = Clientes(id_cliente, nome_cliente, email_cliente, telefone_cliente, interesses_cliente)
                    self.clientes.append(instancia_clientes)
                print("Base de dados de clientes carregada com sucesso!")
        else:
            self.salvar_dados_clientes("locadora.json")
            print("Arquivo 'locadora.json' criado com sucesso.")

    # Passar o nome do arquivo para atualizar o JSON ("locadora.json")
    def salvar_dados_clientes(self, nome_arquivo): 
        # Converte a lista de objetos Clientes para uma lista de dicionários
        dados_para_salvar = [
            {
                'id': c.id_cliente,
                'nome': c.nome_cliente,
                'email': c.email_cliente,
                'telefone': c.telefone_cliente,
                'interesses': c.interesses_cliente
            }
            for c in self.clientes
        ]
        with open(nome_arquivo, 'w', encoding='utf-8') as arquivo_json:
            json.dump({'clientes': dados_para_salvar}, arquivo_json, indent=4)
        print(f"Dados atualizados e salvos em '{nome_arquivo}'.")

    def atualizar_dados(self, id_cliente):
        cliente_encontrado = None
        for cliente in self.clientes:
            # Assumindo que o atributo é 'id_cliente' conforme seu código
            if cliente.id_cliente == id_cliente:
                cliente_encontrado = cliente
                break 
        if cliente_encontrado:
            print(f"Dados atuais de {cliente_encontrado.nome_cliente}:")
            print(f"Nome: {cliente_encontrado.nome_cliente}, E-mail: {cliente_encontrado.email_cliente}")
            nome_cliente = input("Informe o novo nome do cliente: ")
            email_cliente = input("Informe o novo e-mail do cliente: ")
            telefone_cliente = input("Informe o novo telefone do cliente: ")
            interesses_input = input("Informe os interesses separados por vírgula: ")
            lista_interesse_cliente = [item.strip() for item in interesses_input.split(',')]
            # Chama o método de atualização do objeto cliente na memória
            cliente_encontrado.cliente_atualizar_info(id_cliente, nome_cliente, email_cliente, telefone_cliente, lista_interesse_cliente)
            # Após a atualização do objeto, salve os dados no arquivo
            self.salvar_dados_clientes('locadora.json') 
        else:
            print(f"Cliente com ID {id_cliente} não encontrado.")

    def apagar_cliente(self, id_cliente):
        cliente_encontrado = None
        for cliente in self.clientes:
            # Assumindo que o atributo é 'id_cliente' conforme seu código
            print(cliente.id_cliente)
            if cliente.id_cliente == id_cliente:
                cliente_encontrado = cliente
                print(f"Cliente encontrado: {cliente_encontrado}")
                break

        if cliente_encontrado:
            self.clientes.remove(cliente_encontrado)
            self.salvar_dados_clientes('locadora.json')
                
    def cadastrar_clientes(self, nome_cliente, email_cliente, telefone_cliente, interesses_cliente):
        id_cliente = len(self.clientes) + 1
        cliente = Clientes(id_cliente, nome_cliente, email_cliente, telefone_cliente, interesses_cliente)
        self.clientes.append(cliente)
        self.salvar_dados_clientes("locadora.json")
        print()
        print(f"Cliente {cliente.get_nome()} Cadastrado com sucesso")
        
    # def atualizar_dados(self, id_cliente):

    #     for cliente in self.clientes:
    #         if cliente.id_cliente == id_cliente:
    #             nome_cliente = input("Informe o novo nome do cliente.")
    #             email_cliente = input("Informe o novo e-mail do cliente")
    #             telefone_cliente = input("Informe o novo telefone do cliente")
    #             interesses_cliente = input("Informe os interesses separados por virgula.")
    #             lista_interesse_cliente = interesses_cliente.split(",")                
    #             cliente.cliente_atualizar_info(id_cliente, nome_cliente, email_cliente, telefone_cliente, lista_interesse_cliente)
    
    def listar_interesses_cliente(self, nome_cliente=None):
        for cliente in self.clientes:
            if not nome_cliente  == None:
                if cliente.nome_cliente == nome_cliente:
                    cliente.get_interesses(nome_cliente)
                    break
            else:
                cliente.get_interesses(nome_cliente)
    
    def consultar_dados(self):        
        for cliente in self.clientes:
            cliente.get_info_clientes()
```

```python
def kwords(**kwargs):
    for key, value in kwargs.items():
        print(f"Key: {key} - Value: {value}")

def trata_texto(texto):
    quebra_texto = texto.split(",")
    nova_lista = []
    dicionario = []
    
    for txt in quebra_texto:
        dicionario_com_indice = txt.strip()
        nova_lista.append(dicionario_com_indice)
        
    for key, value in enumerate(nova_lista):
        dicionario.append({key:value})
    
    return kwords(dicionario)
    
    
valores = input("Digite os valores separados por virgula.")
valor = trata_texto(valores)
print(valor)
```
